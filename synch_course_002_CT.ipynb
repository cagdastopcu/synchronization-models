```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Euler(f, t, x0):
  dim = len(x0)
  x = np.zeros((len(t),dim))
  x[0] = x0
  for idx in range(len(t)-1):
    h = t[idx+1] - t[idx]
    x[idx+1] = x[idx] + h * f(t[idx],x[idx])
  return x
```

```python
def xdot(t,x): return 2*x
x0 = [1]
t_init = 0.0
t_final = 4.0
h = 0.05
t = np.arange(t_init,t_final,h)
#print(t)
x_euler = Euler(xdot,t,x0)
x_exact = np.exp(2*t)
#print(x_euler)
```

```python
plt.figure(figsize= (10,5))
plt.plot(t,x_euler,'--bx')
plt.plot(t,x_exact,'--kx')

plt.legend()
plt.show()
```
